```python
#importing required libraries 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#Reading data set and X
dataset = pd.read_csv('Iris.csv')
X = dataset.iloc[:,:-1].values

X.shape
```

```python
#applying model for n clusters and extracting inertia
ilist = []
n = 5

for i in range(1,n):
     kmeans = KMeans(n_clusters = i)
     kmeans.fit(X)
     ilist.append(kmeans.inertia_)
```

```python
#plotting elbow graph to predict the optimum number of clusters 
plt.plot(range(1,n), ilist)
plt.title('Elbow')
plt.xlabel('clusters')
plt.ylabel('inertias')
plt.show()
```

```python
#applying model for 3 clusters 
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualising the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 10, c = 'r')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 10, c = 'b')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 10, c = 'g')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'y')
plt.title('Clusters of Species')
plt.xlabel('sepal_x')
plt.ylabel('Sepal_y')
plt.legend()
plt.show()
```
